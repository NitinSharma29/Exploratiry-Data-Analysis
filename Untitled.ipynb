# 2. Perform a data audit for the datatypes and find out if there are any mismatch within the current datatypes of the columns and their business significance.

```python
cust360["DateOfBirth"] = pd.to_datetime(cust360.DateOfBirth, format = "%d-%b-%y")
cust360.loc[(cust360.DateOfBirth.dt.year > 2020),"DateOfBirth"]=cust360[cust360.DateOfBirth.dt.year > 2020]["DateOfBirth"].apply(lambda x: x - pd.DateOffset(years=100))
cust360["claim_date"] = pd.to_datetime(cust360.claim_date, format = "%m/%d/%Y")
```

### 3. Convert the column claim_amount to numeric. Use the appropriate modules/attributes to remove the '$' sign

```python
cust360['Contact'] = pd.to_numeric(cust360.Contact,downcast='float', errors='ignore')
```

```python
cust360['claim_amount']=(cust360['claim_amount']).str.replace('$','')
```

```python
cust360['claim_amount']=pd.to_numeric(cust360.claim_amount,downcast='float', errors='ignore')
```

# 4. Of all the injury claims, some of them have gone unreported with the police. Create an alert flag (1,0) for all such claims.

```python
cust360["flag"] =np.where(cust360.police_report=='No',0, np.where(cust360.police_report=='Yes','Unknown',1))
```

## 5. One customer can claim for insurance more than once and in each claim, multiple categories of claims can be involved. However, customer ID should remain unique.

```python
cust360.drop_duplicates(subset=['customer_id'],inplace=True)
```

```python
cust360.head()
```

## 6. Check for missing values and impute the missing values with an appropriate value. (mean for continuous and mode for categorical)

```python
categorial_col=['CUST_ID','gender','DateOfBirth','State','Segment','claim_id','customer_id','incident_cause','claim_area','police_report','claim_type','fraudulent','flag']
continous_col=['claim_amount','Contact',]
```

```python
for col in categorial_col:
    cust360[col]=cust360[col].fillna(cust360[col].mode()[0])
    cust360[continous_col]=cust360[continous_col].fillna(cust360[continous_col].mean())
```
